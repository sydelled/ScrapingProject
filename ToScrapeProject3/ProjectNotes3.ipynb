```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/selledge/Documents/pythonproject/ScrapingProject-main/book_data.csv')
```

```python
#change columns to upper case
df.columns = [x.upper() for x in df.columns]

#change availability row to lower case
df['AVAILABILITY'] = df['AVAILABILITY'].str.lower()

#change star rating string row
df['RATING']= df['RATING'].replace({'star-rating One': 'one','star-rating Two': 'two','star-rating Three': 'three','star-rating Four': 'four','star-rating Five': 'five'})

#rename price column to price GBP
df.rename(columns = {'PRICE': 'PRICE GBP'}, inplace=True)

#add exchange rate row
df['EXCHANGE RATE']=1.1451

#change british pounds to usd dollars
def price_usd(currency):


    exchange_rate = 1.1451

    currency = currency / exchange_rate

    return round(currency, 2)

#creates new row with usd price
df['PRICE USD'] = df['PRICE GBP'].apply(price_usd)
```

```python
#drops data over 45.00 dollars
filt = df['PRICE USD'] > 45.00
df.drop(index=df[filt].index)
```

```python
#ascend by alphabtical order or if numerical, by smallest to highest
df.sort_values(by='RATING', ascending=False, inplace=True)

```
